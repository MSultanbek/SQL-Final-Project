# Implementing the AI-Agent

Implementing the AI-Agent for our hotel management system

### Installing the agent

```python
from langchain_groq import ChatGroq
from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits import create_sql_agent
from sqlalchemy import create_engine
```

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key="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",
    temperature=0
)
```

```python
engine = create_engine(
    "mysql+pymysql://root:g1k8tc@127.0.0.1/hotel_management"
)

db = SQLDatabase(engine)
print(db.get_usable_table_names())
```

### Building the agent

```python
agent = create_sql_agent(
    llm=llm,
    db=db,
    verbose=True,
    handle_parsing_errors=True,
    max_iterations=15,
    max_execution_time=60
)
```

### Testing

```python
response = agent.invoke(
    "Which month had the highest number of bookings across all years?"
)
print(response['output'])
```

```python
questions = [
    "Which year and month had the highest cancellation rate?",
    "Which customer type generates the highest average daily rate?",
    "Do repeated guests make more special requests than new guests?",
    "Which company has made the most bookings?"
]

for q in questions:
    print(f"\nQuestion: {q}")
    response = agent.invoke(q)
    print(f"Answer: {response['output']}")
    print("-" * 50)
```
